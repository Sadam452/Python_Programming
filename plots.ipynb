```python
import random
list1=[]
list2=[]
for i in range(1,11):
    x= random.randint(1,1000)
    y= random.randint(1,500)
    list1.append(x)
    list2.append(y)
```

```python
import matplotlib.pyplot as plt
plt.plot(list2, list1, color = 'red', linewidth = 2, label = 'List1 vs list2')
plt.xlabel('list2')
plt.ylabel('list1')
plt.title('list1 vs list2')
plt.legend()
plt.show()
```

```python
x=[]
y=[]
for i in range(1,10):
    x.append(i)
    y.append(x[i-1]+i)
plt.plot(x,y)
plt.xlabel('list x')
plt.ylabel('list y')
plt.title('Linear Plot')
plt.show()
```
